```python
import cv2
import xml.etree.ElementTree as ET
import os  
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ANNOTATION_SIGN = "./Traffic signs/Annotations/image_864.xml"
ANNOTATION_TXT_WORD_LEVEL = "./Text/Word Level/Annotations/image_864.xml"
ANNOTATION_TXT_LINE_LEVEL = "./Text/Line Level/Annotations/image_864.xml"
IMAGES_PATH = "./Traffic signs/Images/image_864.png"
```

```python
def visualize(image,annotation):


    root = ET.parse(annotation).getroot()
    iterator = root.getiterator("object")
    img = cv2.imread(image)

    for item in iterator:
        bndbox = item.find("bndbox")
        name = item.find("name").text
        xmin = int(bndbox.find("xmin").text)
        ymin = int(bndbox.find("ymin").text)
        xmax = int(bndbox.find("xmax").text)
        ymax = int(bndbox.find("ymax").text)
        
        img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (0, 0, 255), 2)
        #img = cv2.putText(img, name, (xmax+10, ymax-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    plt.figure(figsize = (20,10))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

## TRAFFIC SIGNS


```python
visualize(IMAGES_PATH,ANNOTATION_SIGN)
```

## TEXT WORD LEVEL

```python
visualize(IMAGES_PATH,ANNOTATION_TXT_WORD_LEVEL)
```

## TEXT LINE LEVEL

```python
visualize(IMAGES_PATH,ANNOTATION_TXT_LINE_LEVEL)
```
